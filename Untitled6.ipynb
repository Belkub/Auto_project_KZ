### Цель и задачи исследований
  Исследование сырых данных предполагает выполнение нескольких этапов подготовки данных для выявления практического анализа. 
Подготовка данных включает анализ дубликатов, заполнение пропусков, числовых и категориальных значений, анализ типа данных и
его коррекцию. Задачами исследования является общий анализ автомобильного рынка, включающий выявление общих закономерностей
продаж и их динамику, выявление топовых марок автомобилей и особенностенй сегментации авторынка (по регионам и автоцентрам), а
также устанвление ценовых предпочтений и тенденций. Целью исследования является формулировка основных рекомендаций для 
автодилеров по увеличению объемов продаж на основании оценки общей и сегментированной емкостей рынка, ценовых и брендовых
предпочтений клиентов, а также ожидаемой прибыльности и общей конъюнктуры.

```python
for i in cd:
    minp = dfs1.loc[dfs1['Модель'] == i, 'Цена, USD'].min()
    maxp = dfs1.loc[dfs1['Модель'] == i, 'Цена, USD'].max()
    miny = dfs1.loc[dfs1['Модель'] == i, 'Год выпуска'].min()
    maxy = dfs1.loc[dfs1['Модель'] == i, 'Год выпуска'].max()
    if (maxp - minp) == 0:
        df.loc[(df['Год выпуска'] == 0) & (df['Модель'] == i), 'Год выпуска'] = miny
    if (maxy - miny) == 0:
        df.loc[(df['Год выпуска'] == 0) & (df['Модель'] == i), 'Год выпуска'] = miny
    if (maxy - miny) != 0 and (maxp - minp) != 0:
        k  = (maxy - miny)/(maxp - minp) 
        dfw = dfs.loc[dfs['Модель'] == i] 
        year = round(miny + dfw['Цена, USD'].mean()*k)
        f.append(year)
        if year > 2019:
            year = 2019
        df.loc[(df['Год выпуска'] == 0) & (df['Модель'] == i), 'Год выпуска'] = year
df['Год выпуска'].unique()
```

```python
def gr(w):
    if w == 'FWD':
        return 1
    elif w == 'RWD':
        return 2
    elif w == '2WD':
        return 3
    elif w == '4WD':
        return 4
    else:
        return 0
df['привод'] = df['Тип привода'].apply(gr)
df['привод'].unique()
```

```python
for i in df['Модель'].unique():
    df.loc[(df['Модель'] == i) & (df['Тип привода'] == 'Не определен') & (df['привод'] == 1), 'Тип привода'] = 'FWD'
    df.loc[(df['Модель'] == i) & (df['Тип привода'] == 'Не определен') & (df['привод'] == 2), 'Тип привода'] = 'RWD'
    df.loc[(df['Модель'] == i) & (df['Тип привода'] == 'Не определен') & (df['привод'] == 3), 'Тип привода'] = '2WD'
    df.loc[(df['Модель'] == i) & (df['Тип привода'] == 'Не определен') & (df['привод'] == 4), 'Тип привода'] = '4WD'
```

Для множества моделей с пропусками в столбце 'Год выпуска', которое не пересекается с множеством моделей с известными значениями 
в данном столбце, заполнить пропуски не представляется возможным, поэтому они были обнулены. Данный прием позволил снизить
число неизвестных значений в столбце 'Год выпуска' с 240 до 55.  
    Названия стран-производителей авто были приведены к международной кодировке.
